```python
# Horizontal Bar Chart
```

```python
import matplotlib.pyplot as plt
```

```python
# Data
```

```python
categories = ['A', 'B', 'C']
values = [23, 45, 30]
```

```python
# Plot
```

```python
plt.barh(categories, values, color='pink')
```

```python
plt.title("Horizontal Bar Chart")
```

```python
plt.xlabel("Values")
```

```python
plt.ylabel("Categories")
```

```python
plt.show()
```
