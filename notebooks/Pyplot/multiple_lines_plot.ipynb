```python
#Line Plot with Multiple Lines
```

```python
import matplotlib.pyplot as plt
```

```python
# Data
```

```python
x = [0, 1, 2, 3, 4]
y1 = [0, 1, 4, 9, 16]
y2 = [0, 2, 6, 12, 20]
```

```python
# Plot
```

```python
plt.plot(x, y1, label="Line 1", color="blue", marker="o")
plt.plot(x, y2, label="Line 2", color="red", linestyle="--", marker="x")
```

```python
plt.title("Multiple Lines Plot")
```

```python
plt.xlabel("X-axis")
```

```python
plt.ylabel("Y-axis")
```

```python
plt.legend()
plt.show()
```
