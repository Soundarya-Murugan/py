```python
#Box Plot
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
```

```python
plt.boxplot(data, vert=True, patch_artist=True)
```

```python
plt.title("Box Plot")
```

```python
plt.xlabel("Categories")
```

```python
plt.ylabel("Values")
```

```python
plt.show()
```
