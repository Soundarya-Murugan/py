```python
#Basic Line Plot
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
```

```python
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y, marker='o')
```

```python
plt.title("Basic Line Plot")
```

```python
plt.xlabel("X-axis")
```

```python
plt.ylabel("Y-axis")
```

```python
plt.grid()
```

```python
plt.show()
```
