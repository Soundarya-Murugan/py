```python
import matplotlib.pyplot as plt
```

```python
data = [
    ["Alice", 24, "Engineer"],
    ["Bob", 27, "Data Scientist"],
    ["Charlie", 22, "Teacher"]
]
```

```python
columns = ("Name", "Age", "Occupation")
```

```python
fig, ax = plt.subplots()
```

```python
ax.axis("tight")
```

```python
ax.axis("off")
```

```python
table = ax.table(cellText=data, colLabels=columns, cellLoc="center", loc="center")
```

```python
plt.title("Simple Table Example")
```

```python
plt.show()
```
