```python
#Stacked Bar Chart
```

```python
import matplotlib.pyplot as plt
```

```python
# Data
```

```python
categories = ['A', 'B', 'C']
values1 = [3, 5, 2]
values2 = [4, 6, 3]
```

```python
# Plot
```

```python
plt.bar(categories, values1, label="Dataset 1", color="blue")
```

```python
plt.bar(categories, values2, bottom=values1, label="Dataset 2", color="green")
```

```python
plt.title("Stacked Bar Chart")
```

```python
plt.legend()
plt.show()
```
